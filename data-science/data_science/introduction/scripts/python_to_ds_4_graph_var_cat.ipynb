# working with data #

```python
#compare english and other languages 
count_language =  tmdb.original_language.value_counts() # series with counts 
english_count = count_language.loc['en']
total_language = count_language.sum()
total_language
no_english = (total_language-english_count)
no_english
```

```python
data_to_plot = {
    'language' : ['english', 'others'],
    'values' : [english_count, no_english]
}
data_to_plot
pd.DataFrame(data_to_plot)
sns.barplot(x= 'language', y = 'values', data= data_to_plot)
print ("task")

# more informative plot
```
